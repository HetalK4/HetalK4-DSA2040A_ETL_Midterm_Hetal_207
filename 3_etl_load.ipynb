## Visual

```python

import pandas as pd
import matplotlib.pyplot as plt

raw_data = pd.read_csv("2_transformed/transformed_full.csv")

# Aggregating total sales by product
total_sales = raw_data.groupby('product')['total_price'].sum().reset_index()

# Sorting the data in descending orderfor better visualization
total_sales = total_sales.sort_values(by='total_price', ascending=False)

# Creating a bar chart
plt.figure(figsize=(10, 6))  # Setting the figure size
plt.bar(total_sales['product'], total_sales['total_price'], color='skyblue')
plt.title('Total Sales by Product', fontsize=16)  # Title of the chart
plt.xlabel('Product', fontsize=14)  # X-axis label
plt.ylabel('Total Sales', fontsize=14)  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add grid lines for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
# Displaying the bar chart
plt.show()
```

Insights:

- 'Tablet' have the highest sales whereas 'Phone' has the lowest.
- All products have made sales greater than 15000.
